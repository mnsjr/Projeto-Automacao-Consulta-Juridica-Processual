```python
# criar o navegador
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
# abrir a página index (entrar no site da busca jurídica)
import os

caminho = os.getcwd()
arquivo = caminho + r"\index.html"
```

```python
# importar a base de dados
import pandas as pd

tabela = pd.read_excel("Processos.xlsx")
display(tabela)
```

```python
from selenium.webdriver import ActionChains
import time


# para cada processo (linha da tabela)
for linha in tabela.index:
    navegador.get(arquivo)


    # abrir a lista de cidades
    botao = navegador.find_element(By.CLASS_NAME, 'dropdown-menu')
    ActionChains(navegador).move_to_element(botao).perform()   # colocar o mouse em cima do botão
    time.sleep(0.5)
    
    # selecionando o cidade
    cidades = tabela.loc[linha, 'Cidade']
    # o By.PARTIAL_LINK_TEXT vai procurar no texto do link no html o link/botão com o texto parecido com a variável cidades
    navegador.find_element(By.PARTIAL_LINK_TEXT, cidades).click() 
    
    # mudar para a aba_form, que a cada nova pagina terá um indice diferente
    # posso utilizar o indice da tebela como contador para o idnice da pagina
    time.sleep(0.5)
    aba_original = navegador.window_handles[0]
    indice_aba = 1 + linha  # a variável linha é o indice/contador da tabela no for
    nova_aba = navegador.window_handles[indice_aba]
    
    navegador.switch_to.window(nova_aba)

    
    # preencher o formulário com os dados de busca
    navegador.find_element(By.ID, 'nome').send_keys(tabela.loc[linha, "Nome"])
    navegador.find_element(By.ID, 'advogado').send_keys(tabela.loc[linha, "Advogado"])
    navegador.find_element(By.ID, 'numero').send_keys(tabela.loc[linha, "Processo"])
    
    # clicar em pesquisar
    navegador.find_element(By.CLASS_NAME, 'registerbtn').click()
    
    # confirmar a pesquisa
    alerta = navegador.switch_to.alert
    alerta.accept()

    # esperar o resultado da pesquisa e agir de acordo com o resultado
    # se eu buscar o alerta 'alerta = navegador.switch_to.alert' e ele não estiver na página, aparece um erro
    # ideal seria utilizar o 'expected conditions', mas como neste caso sabemos com certeza que um alerta irá aparecer, só esperar
    # desta forma vou tentar encontrar o alerta, se não encontrar, esperar 1 segundo e tentar novamente
    # quando achar break para sair do loop infinito.
    contador = 0
    while True:
        try:
            alerta = navegador.switch_to.alert
            break
        except:
            time.sleep(1)
            contador += 1
            if contador > 30:
                break 
            
    texto_alerta = alerta.text

    if 'Processo encontrado com sucesso' in texto_alerta:
        alerta.accept()
        tabela.loc[linha, 'Status'] = 'Encontrado'
        print('Processo encontrado.')
    else:
        alerta.accept()
        tabela.loc[linha, 'Status'] = 'Não encontrado'
        print('Processo não encontrado.')
    
# fechando a abaaba_formulario
navegador.quit()
```

```python
# exibindo tabela atualizada
display(tabela)

# exportando a tabela atualizada em formato excel
tabela.to_excel('Processos Atualizados.xlsx')
```
